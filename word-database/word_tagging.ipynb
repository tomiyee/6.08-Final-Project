```python
from rake_nltk import Rake
import nltk
import pandas as pd
import re
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_colwidth', 500)
pd.set_option('display.width', 1000)

r = Rake(max_length = 1)

dict_file = "raw_full_dictionary.txt"
df = pd.read_csv(dict_file, sep = "=", names = ["word", "meaning"])
filtered = df.copy()
```

```python
def meaning_filter(x):
    first_def = x.split(";")[0]
    first_def = re.sub(r'.* pertaining to', 'pertaining to', first_def)
    first_def = re.sub('or resembling', '', first_def)
    return first_def
```

```python
filtered = df.copy()
filtered['meaning'] = filtered['meaning'].apply(meaning_filter)

filtered['num_words'] = filtered['meaning'].str.split().str.len()
filtered = filtered[~filtered['meaning'].str.contains('-')]
filtered = filtered[~filtered['meaning'].str.contains(' or ')]
short = filtered[filtered['num_words'] <= 3]
short
```

```python
total = 0
best_dict = dict()
output = ""

for i, row in short.iterrows():
    word, meaning, num = row
    try:
        r.extract_keywords_from_text(meaning)
        best = r.get_ranked_phrases()[0]
        if best in ['pertaining', 'science', 'chemistry', 'study', 'making', 'writing', 'love', 'fear', 'resembling']:
            best = r.get_ranked_phrases()[1]
            if best in best_dict:
                best_dict[best] += 1
            else:
                best_dict[best] = 1
        print("{}={}={}".format(word, meaning.replace(best, "_____", 1),best))
        output += "{}={}={}\n".format(word, meaning.replace(best, "_____", 1),best)
        total += 1
    except:
        pass
with open("bluffalo_words_with_blank.txt", "w", encoding='utf-8') as out:
    out.write(output)
print(total)
```
